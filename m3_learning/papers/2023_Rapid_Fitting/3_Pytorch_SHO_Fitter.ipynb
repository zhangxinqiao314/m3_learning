```python
BE_viz = Viz(dataset, printing, verbose=True)


state = {'fitter' : 'LSQF',
             'resampled' : True,
             'scaled' : True,
             "label": "Scaled"}

BE_viz.nn_checker(state, filename="Figure_8_Scaled Raw Data")
```
